```python
df_mart['연령대'].value_counts()
```

```python
df_mart['연령대'] == '40대'
```

```python
age_40 = (df_mart['연령대'] == '40대')
age_40
```

```python
df_mart.loc[age_40] # 40대만 추출
```

```python
# 특정 데이터만 추출하는 방법, 한줄로 처리
df_mart.loc[df_mart['연령대'] == '40대']
```

```python
# 40대를 제외하고 추출하는 함수.
df_mart.loc[df_mart['연령대'] != '40대']
```

```python
# 40대이면서 남성인 데이터만 추출.
df_mart.loc[(df_mart['연령대'] == '40대') & (df_mart['성별']=='남')]
```

```python
# 컬럼 정보를 리스트로 가져오기
df_mart.describe().columns.tolist()
```

```python
# 문자형데이터는 모두 제거 하는 경우
total_list = df_mart.columns.tolist()
total_list
```

```python
#  문자정보를 뽑아낸것
num_list = df_mart.describe().columns.tolist()
tmp_list = list(set(total_list) - set(num_list))
df_mart[tmp_list]
```

```python
# 숫자를 뽑아내고 싶은 경우
df_mart[df_mart.describe().columns.tolist()]
```
